```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("./basic_test/result.csv")

i = "loss_on_full_trainset"
j = "gradient_norm_on_full_trainset"
end = 6000

plt.plot(df1[i][:end], label="loss_on_full_trainset")
plt.plot(df1[j][:end], label="gradient_norm_on_full_trainset")
plt.title("rho=0,uniform data,lr=0.8")
plt.yscale("log")
plt.grid()
plt.legend()
plt.show()
```
