```python
def ring1(n=10):  # 生成稀疏环状图。也可以取n=5
    A, B = np.eye(n) / 2, np.eye(n) / 2
    m = int(n / 2)
    for i in range(n - 1):
        A[i][i + 1] = 0.5
        B[i][i + 1] = 0.5
    A[n - 1][0] = 0.5
    B[n - 1][0] = 0.5
    A[0][m] = 1 / 3
    A[m - 1][m] = 1 / 3
    A[m][m] = 1 / 3
    B[0][0] = 1 / 3
    B[0][1] = 1 / 3
    B[0][m] = 1 / 3
    return A.T, B.T  # A.T是行随机，B.T是列随机矩阵，
```

```python
import numpy as np
import matplotlib.pyplot as plt

def prettyshow(grads, legend, save='image.pdf', ylabel='Gradient Norm', title=None):
    # plot the results
    plt.rcParams['figure.figsize'] = 5, 4
    plt.figure()
    xlen = len(grads[0])
    colors = ['green', 'red', 'blue', 'orange', 'purple', 'cyan']
    markers = ['d', '^', 'o', '<', '*', 's']
    idx_set = np.arange(0, xlen, xlen//10)
    
    for i in range(len(grads)):
        plt.semilogy(0, grads[i][0], color=colors[i], marker=markers[i], markersize=7)
    
    for i in range(len(grads)):
        for idx in idx_set:
            plt.semilogy(idx, grads[i][idx], color=colors[i], marker=markers[i], markersize=7, linestyle='None')
    
    for i in range(len(grads)):
        plt.semilogy(np.arange(xlen), grads[i], linewidth=1.0, color=colors[i])
    
    # 将图例放置在右上角
    plt.legend(legend, fontsize=12, loc='upper right')
    
    # 添加标题（如果提供了 title 参数）
    if title:
        plt.title(title, fontsize=14)
    
    plt.xlabel('Iteration', fontsize=12)
    plt.ylabel(ylabel, fontsize=12)
    plt.grid(True)

    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    
    # Adjust the left margin to prevent y-label from being cut off
    plt.subplots_adjust(left=0.15)  # Increase the left margin
    
    plt.savefig(save)
    plt.show()
```

```python
df_combined = pd.read_csv('multigossip.csv')

a = 7500  # 假设 a 是所有的索引

# 提取列数据
l5 = df_combined['l5'].values[:a]
s6 = df_combined['s6'].values[:a]
t6 = df_combined['t6'].values[:a]
p9 = df_combined['p9'].values[:a]

# 将数据放入一个列表中
grads = [l5, s6, t6, p9]

# 定义图例标签
legend = ['lr=6e-3, MG=1', 'lr=7e-3, MG=2', 'lr=7e-3, MG=5', 'lr=1e-2, MG=10']

# 调用 prettyshow 函数
prettyshow(grads, legend, save='Multi_Gossip_合成数据_PullDiag_GT.pdf', ylabel='Gradient Norm', title='Ring, Logistic')
```
