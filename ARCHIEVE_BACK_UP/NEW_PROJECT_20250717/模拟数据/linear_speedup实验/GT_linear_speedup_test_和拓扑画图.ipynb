```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 定义移动平均窗口大小（根据数据调整）
window_size = 100

# 数据集和标签
data_list = [
    L1["gradient_norm_on_full_trainset"],
    #L2["gradient_norm_on_full_trainset"],
    L3["gradient_norm_on_full_trainset"],
    #L4["gradient_norm_on_full_trainset"],
    L5["gradient_norm_on_full_trainset"],
    L6["gradient_norm_on_full_trainset"],
]
labels = [
    "n=1",
    #"n=2",
    "n=8",
    #"n=16",
    "n=128",
    "n=512",
]

plt.figure()

for data, label in zip(data_list, labels):
    s = pd.Series(data)
    # 计算移动平均和标准差
    ma = s.rolling(window=window_size, center=True).mean()
    std = s.rolling(window=window_size, center=True).std()
    # 计算上下边界
    upper = ma + std
    lower = ma - std
    # 去除NaN值
    valid = ~ma.isna()
    x = np.arange(len(data))[valid]
    # 绘制均值实线
    (line,) = plt.semilogy(x, ma[valid], label=f"FULL_gradient_norm, {label}")
    color = line.get_color()
    # 填充震荡区域（均值±标准差）
    plt.fill_between(x, lower[valid], upper[valid], color=color, alpha=0.2)

plt.xlabel("Training Steps")
plt.ylabel("Gradient Norm (log scale)")
plt.legend()

# 保存为 PDF 文件
plt.savefig("/root/GanLuo/ICML2025_project/模拟数据/linear_speedup实验/output/GD_指数图_没有重复实验_1.pdf", format="pdf", bbox_inches="tight")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker

import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker

def prettyshow(grads, legend, save='image.pdf', ylabel='Gradient Norm'):
    plt.rcParams['figure.figsize'] = 12, 8
    
    plt.figure()
    xlen = len(grads[0])
    colors = ['green', 'red', 'blue', 'orange', 'purple', 'brown']
    markers = ['d', '^', 'o', '<', '*', 's']
    idx_set = np.arange(0, xlen, xlen//10)
    
    for i in range(len(grads)):
        plt.semilogy(0, grads[i][0], color=colors[i], marker=markers[i], markersize=8)
    
    for i in range(len(grads)):
        for idx in idx_set:
            plt.semilogy(idx, grads[i][idx], color=colors[i], marker=markers[i], markersize=15, linestyle='None')
    
    for i in range(len(grads)):
        plt.semilogy(np.arange(xlen), grads[i], linewidth=4, color=colors[i])
    
    # 设置图例为2行3列，并放置在右上角
    plt.legend(legend, prop={'size': 30, 'weight': 'bold'}, loc='upper left', 
               bbox_to_anchor=(0.03, 1), ncol=2)
    
    plt.xlabel('Comm. Rounds', fontsize=47, fontweight='bold')
    plt.ylabel(ylabel, fontsize=47, fontweight='bold')
    
    # 设置科学记数法格式
    ax = plt.gca()
    formatter = ticker.ScalarFormatter(useMathText=True)
    formatter.set_scientific(False)
    formatter.set_powerlimits((3, 6))  # 当数值超过1e3或1e6时使用科学记数法
    ax.xaxis.set_major_formatter(formatter)
    
    # 自定义x_tick显示，将数值除以100
    def format_x_tick(x, pos):
        return f'{x/1:.0f}'
    
    ax.xaxis.set_major_formatter(ticker.FuncFormatter(format_x_tick))
    
    # 调整刻度数量和间隔
    ax.xaxis.set_major_locator(ticker.MaxNLocator(nbins=5))
    
    # 设置刻度标签和偏移文本的字体
    plt.xticks(fontsize=35, fontweight='bold')
    plt.yticks(fontsize=35, fontweight='bold')
    ax.xaxis.get_offset_text().set_fontsize(25)  # 调整科学记数法指数字体大小
    ax.xaxis.get_offset_text().set_fontweight('bold')
    plt.title("Exponential", fontsize=47, fontweight='bold')
    
    plt.grid(True)
    plt.subplots_adjust(left=0.21, bottom=0.15, right=0.75)  # 调整右侧边距以容纳图例
    plt.savefig(save, bbox_inches='tight')
    plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

def draw_ring_graph(n, extra_edges, ax, title):
    G = nx.DiGraph()
    nodes = list(range(n))
    
    # Add nodes
    G.add_nodes_from(nodes)
    
    # Add ring edges (cyclic)
    for i in range(n):
        G.add_edge(nodes[i], nodes[(i + 1) % n])
    
    # Add extra edges
    G.add_edges_from(extra_edges)

    # Circular layout
    pos = nx.circular_layout(G)
    
    # Draw nodes and edges
    nx.draw_networkx_nodes(G, pos, node_color="lightblue", edgecolors="gray", node_size=800)
    nx.draw_networkx_labels(G, pos, font_size=10, font_weight="bold")
    nx.draw_networkx_edges(G, pos, edgelist=G.edges(), width=2, edge_color="gray", arrows=True)
    
    # Title
    ax.set_title(title, fontsize=14, fontstyle='italic')
    ax.axis("off")

# Create the figure
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Define the number of nodes
num_nodes = 20

# Define the additional edges for the two cases
extra_edges_ring3 = [(2, 10), (10, 15), (2, 15)]
extra_edges_ring4 = [(0, 5), (5, 15), (0, 15), (10, 5)]

# Draw both graphs
draw_ring_graph(num_nodes, extra_edges_ring3, axes[0], r"$Ring_3$")
draw_ring_graph(num_nodes, extra_edges_ring4, axes[1], r"$Ring_4$")

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

def generate_exponential_weight_matrix(n):
    """
    生成静态指数图的权重矩阵。

    参数：
    n (int): 图中节点的数量。

    返回：
    numpy.ndarray: 形状为(n, n)的权重矩阵，元素类型为float。
    """
    if n < 1:
        raise ValueError("n必须为正整数")
    
    # 计算分母：|log2(n)| + 1
    denominator = np.abs(np.log2(n)) + 1
    
    # 创建索引网格
    i_indices, j_indices = np.indices((n, n))
    
    # 计算mod_val = (j - i) mod n
    mod_vals = (j_indices - i_indices) % n
    
    # 判断是否为2的幂
    is_power_of_two = (mod_vals != 0) & ((mod_vals & (mod_vals - 1)) == 0)
    
    # 判断是否为自环或mod_val是2的幂
    mask = (i_indices == j_indices) | is_power_of_two
    
    # 生成权重矩阵
    weight_matrix = np.where(mask, 1.0 / denominator, 0.0)
    
    return weight_matrix

def plot_exponential_graph(n, filename="exponential_graph.pdf"):
    A = generate_exponential_weight_matrix(n)
    G = nx.DiGraph()

    # 添加节点
    G.add_nodes_from(range(n))

    # 添加边
    for i in range(n):
        for j in range(n):
            if A[i, j] > 0:
                G.add_edge(i, j)

    # 采用环形布局
    pos = nx.circular_layout(G)
    
    # 绘制图形
    plt.figure(figsize=(8, 8))
    nx.draw(G, pos, with_labels=True, node_size=700, node_color="lightblue", edge_color="gray", 
            arrows=True, arrowsize=20, width=2.5, font_weight="bold")
    
    #plt.title(f"Exponential (n={n})", fontsize=20, fontstyle='italic')
    
    # 保存为PDF
    plt.savefig(filename, format="pdf", bbox_inches="tight")
    plt.show()

# 画出 n=16 时的图并保存为 PDF
plot_exponential_graph(8, "exponential_graph_n8(数字，无标题).pdf")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

def draw_grid_graph(rows, cols, filename="grid_graph.pdf"):
    G = nx.DiGraph()

    # 添加节点
    for i in range(rows):
        for j in range(cols):
            G.add_node((i, j))

    # 添加横向和纵向的双向边
    for i in range(rows):
        for j in range(cols):
            if j < cols - 1:
                G.add_edge((i, j), (i, j + 1))
                G.add_edge((i, j + 1), (i, j))
            if i < rows - 1:
                G.add_edge((i, j), (i + 1, j))
                G.add_edge((i + 1, j), (i, j))

    # 采用网格布局
    pos = {(i, j): (j, -i) for i, j in G.nodes()}

    plt.figure(figsize=(6, 6))
    nx.draw(G, pos, with_labels=False, node_size=600, node_color="lightblue", edge_color="gray",
            arrows=True, arrowsize=20, width=2.5, connectionstyle="arc3,rad=0.1")

    plt.title(f"Grid (n=16)", fontsize=20, fontstyle='italic')

    # 保存为PDF
    plt.savefig(filename, format="pdf", bbox_inches="tight")
    plt.show()

# 画出4x4的网格图并保存为PDF
draw_grid_graph(4, 4, "grid_graph_4x4.pdf")
```

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

def ring1(n=10):
    """
    生成稀疏环状图的权重矩阵 A.T 和 B.T。

    参数：
    n (int): 图中节点数量。

    返回：
    tuple: A.T 和 B.T 矩阵
    """
    A, B = np.eye(n) / 2, np.eye(n) / 2
    m = int(n / 2)
    for i in range(n - 1):
        A[i][i + 1] = 0.5
        B[i][i + 1] = 0.5
    A[n - 1][0] = 0.5
    B[n - 1][0] = 0.5
    A[0][m] = 1 / 3
    A[m - 1][m] = 1 / 3
    A[m][m] = 1 / 3
    B[0][0] = 1 / 3
    B[0][1] = 1 / 3
    B[0][m] = 1 / 3
    return A.T, B.T  # A.T 是行随机，B.T 是列随机矩阵

def plot_ring_graph(n=16):
    A_T, _ = ring1(n)

    G = nx.DiGraph()

    # 添加节点
    G.add_nodes_from(range(n))

    # 添加边，根据矩阵 A.T 生成
    for i in range(n):
        for j in range(n):
            if A_T[i, j] > 0:
                G.add_edge(i, j, weight=A_T[i, j])

    # 环形布局
    pos = nx.circular_layout(G)

    plt.figure(figsize=(8, 8))
    nx.draw(G, pos, with_labels=True, node_size=700, node_color="lightblue", edge_color="gray",
            arrows=True, arrowsize=20, width=2.5, font_weight="bold")
    
    # 保存为PDF
    #plt.title(f"Ring (n=16)", fontsize=20, fontstyle='italic')
    plt.savefig("Ring 16(自环+数字，无标题).pdf", format="pdf", bbox_inches="tight")
    plt.show()

# 绘制 n=16 时的环状图
plot_ring_graph(16)
```

```python
def draw_grid_graph_labeled_zero_based(rows, cols, filename="grid_16_labeled_zero_based.pdf"):
    G = nx.DiGraph()
    
    # 生成节点标签，按从上到下、从左到右的顺序编号，从0开始
    node_labels = {}
    count = 0
    for i in range(rows):
        for j in range(cols):
            G.add_node((i, j))
            G.add_edge((i, j), (i, j))  # 添加自环
            node_labels[(i, j)] = count
            count += 1

    # 添加横向和纵向的双向边
    for i in range(rows):
        for j in range(cols):
            if j < cols - 1:
                G.add_edge((i, j), (i, j + 1))
                G.add_edge((i, j + 1), (i, j))
            if i < rows - 1:
                G.add_edge((i, j), (i + 1, j))
                G.add_edge((i + 1, j), (i, j))

    # 采用网格布局，节点按照 (j, -i) 排布，保持图像直观
    pos = {(i, j): (j, -i) for i, j in G.nodes()}

    plt.figure(figsize=(6, 6))
    nx.draw(G, pos, with_labels=True, labels=node_labels, node_size=600, node_color="lightblue",
            edge_color="gray", arrows=True, arrowsize=20, width=2.5, connectionstyle="arc3,rad=0.1")
    
    #plt.title(f"Grid with Labels (0 to {rows * cols - 1})", fontsize=20, fontstyle='italic')

    # 保存为PDF
    plt.savefig(filename, format="pdf", bbox_inches="tight")
    plt.show()

# 画出4x4的网格图并保存为PDF
draw_grid_graph_labeled_zero_based(4, 4, "grid16,new.pdf")
```
